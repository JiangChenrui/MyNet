```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

* cv2.IMREAD_COLOR(): 读入彩色图像
* cv2.IMREAD_GRAYSCALE(): 以灰度模式读入图像

```python
img = cv2.imread('Pictures/壁纸1.jpg', cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
def salt(img, n):
    for k in range(n):
        i = int(np.random.random() * img.shape[1])
        j = int(np.random.random() * img.shape[0])
        if img.ndim == 2:
            img[j, i] = 255
        elif img.ndim == 3:
            img[j,i,0] = 255
            img[j,i,1] = 255
            img[j,i,2] = 255
    return img
```

```python
plt.imshow(salt(img, 10000))
plt.show()
```

```python
imgobj1 = cv2.imread('Pictures/壁纸10.jpg')
imgobj2 = cv2.imread('Pictures/壁纸11.jpg')
hist1 = cv2.calcHist([imgobj1], [0], None, [256], [0.0, 255.0])
hist2 = cv2.calcHist([imgobj2], [0], None, [256], [0.0, 255.0])
plt.plot(range(256), hist1, 'r')
plt.plot(range(256), hist2, 'b')
plt.show()
```

```python
img = cv2.imread('Pictures/壁纸20.jpg')
img_500x500 = cv2.resize(img, (200, 200))
img_500x600 = cv2.resize(img, (0, 0), fx=0.5, fy=0.5,
                        interpolation=cv2.INTER_NEAREST)

img_600x600 = cv2.copyMakeBorder(img, 50, 50, 0, 0,
                                cv2.BORDER_CONSTANT,
                                value=(0, 0, 0))
path_tree = img[20:150, -180:-50]

cv2.imwrite('cropped_tree.jpg', path_tree)
cv2.imwrite('resized_500x500.jpg', img_500x500)
cv2.imwrite('resized_500x600.jpg', img_500x600)
cv2.imwrite('bordered_600x600.jpg', img_600x600)
```
