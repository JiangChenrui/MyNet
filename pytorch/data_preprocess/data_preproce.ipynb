```python
import glob
import cv2
import sys
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
from torchvision import transforms
```

```python
img = cv2.imread('yes_daocha.jpg')
# img = Image.open('yes_daocha.jpg')
print(img[0][0][0])
print(img[0][0][1])
print(img[0][0][2])
plt.imshow(img)
plt.show()
```

```python
img_gray = cv2.imread('yes_daocha.jpg', cv2.IMREAD_GRAYSCALE)
print(img_gray.size)
plt.imshow(img_gray, cmap='gray')
plt.show()
```

```python
ret, binary = cv2.threshold(img_gray,220,255,cv2.THRESH_BINARY)

cv2.imwrite("binary.jpg",binary)
# 
contours, hierarchy = cv2.findContours(binary, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# 
cv2.drawContours(img_gray, contours, -1, (0,0,255), 3)
plt.imshow(img_gray, cmap='gray')
plt.show()

# 

cv2.putText(img_gray, "{:.3f}".format(len ( contours )), (30, 30),cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0,255,0), 1) 
cv2.imwrite('contours.jpg', img_gray)
```
